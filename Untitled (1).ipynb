```python
# Suppression des valeurs aberrantes uniquement pour 'annee'
def remove_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Garder uniquement les valeurs dans l'intervalle
    df = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
    
    return df

# Appliquer le nettoyage uniquement sur 'annee'
df_reel = remove_outliers_iqr(df_reel, 'annee')

print(f"Nombre de lignes après nettoyage : {len(df_reel)}")
```

```python
import pandas as pd
import numpy as np
import joblib
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# -----------------------------
# 1. Charger ton dataset
# -----------------------------
# df = pd.read_csv("ton_dataset.csv")  # si tu as un fichier CSV
# Sinon, utiliser df déjà créé dans ton code

# -----------------------------
# 2. Nettoyer les valeurs infinies
# -----------------------------
df.replace([np.inf, -np.inf], np.nan, inplace=True)

# Remplir les NaN avec la médiane des colonnes numériques
numeric_cols = df.select_dtypes(include=[np.number]).columns
df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].median())

# -----------------------------
# 3. Encodage des colonnes texte
# -----------------------------
cat_cols = df.select_dtypes(include=['object']).columns
for col in cat_cols:
    df[col + "_encoded"] = df[col].astype('category').cat.codes
df.drop(columns=cat_cols, inplace=True)

# -----------------------------
# 4. Créer une cible fictive
# -----------------------------
# Ici on prend 'importance' : si importance > mediane → 1, sinon 0
df['importance_encoded'] = (df['importance'] > df['importance'].median()).astype(int)

# -----------------------------
# 5. Préparer X et y
# -----------------------------
X = df.drop(columns=['importance_encoded'])
y = df['importance_encoded']

# -----------------------------
# 6. Standardisation
# -----------------------------
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# -----------------------------
# 7. Split train/test
# -----------------------------
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# -----------------------------
# 8. Entraînement RandomForest
# -----------------------------
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)

# -----------------------------
# 9. Évaluation
# -----------------------------
y_pred = rf_model.predict(X_test)
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))

# -----------------------------
# 10. Sauvegarde modèle et scaler
# -----------------------------
joblib.dump(rf_model, "bourguiba_rf_model.pkl")
joblib.dump(scaler, "bourguiba_scaler.pkl")
print("\nModèle et scaler sauvegardés avec succès !")
```

```python
import pandas as pd

# Exemple de dataset
df = pd.read_csv("bourguiba_historique_etendu.csv")  # ton dataset historique

# Créer une colonne combinée avec toutes les informations utiles
df['texte'] = df['evenement'] + " " + df['description'] + " " + df['lieu'].astype(str) + " " + df['annee'].astype(str)
corpus = df['texte'].tolist()

```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer()
X_vectors = vectorizer.fit_transform(corpus)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

def repondre_question(question):
    question_vec = vectorizer.transform([question])
    similitudes = cosine_similarity(question_vec, X_vectors)
    idx_max = similitudes.argmax()
    return df.iloc[idx_max]['texte']

# Exemple
question = "Quand Bourguiba a-t-il obtenu son baccalauréat ?"
reponse = repondre_question(question)
print("Réponse :", reponse)
```
